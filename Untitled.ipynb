```python
import numpy as np
import matplotlib.pyplot as plt

def restriccion1(x1):
    return (24 - 6 * x1) / 4.


def z(x1):
    return (11.9 - 3 * x1) / 2.

plt.figure(figsize=(10, 6))

x1 = np.arange(0.0, 3.1, 0.1)

plt.plot(x1, restriccion1(x1), linewidth=1.5)

plt.plot(x1, z(x1), linewidth=1.5)

plt.xlim(0, 3)
plt.ylim(0, 7)
plt.show()
```
