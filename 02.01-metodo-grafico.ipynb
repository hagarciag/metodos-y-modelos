La solución gráfica de un problema de programación lineal permite comprender los tipos de soluciones que se pueden obtener. Considere el siguiente problema:

Función objetivo:

$$\text{max } 0.3x_1 + x_2$$

Sujeto a:
$$ \begin{array}{cccc}
    x_1 & -x_2 & \le & 2\\
   -x_1 & +x_2 & \le & 1\\
    x_1 &      & \ge & 0\\
        &  x_2 & \ge & 0
  \end{array}$$


![alt text](images/pl1.png)

La gráfica anterior fue creada usando la utilidad grapher de macOS. Con el lenguaje Python se procede de la siguiente manera: Se despeja $x_2$  y se escribe la función correspondiente:

```python
# restricción 1
def r1(x1):
    return 2-x1

# restricción 2
def r2(x1):
    return 1+x1
```

Para graficar la función objetivo, se suponen valores diferentes de $z$ y se depeja $x_2$:


```python
def z1(x1):
    return 1.9 - 0.3*x1

def z2(x1):
    return 1.8 - 0.3*x1

def z3(x1):
    return 1.7 - 0.3*x1

def z4(x1):
    return 1.6 - 0.3*x1

def z5(x1):
    return 1.5 - 0.3*x1
```

Y se grafica.

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

x1 = np.arange(0.0, 2.1, 0.1)
plt.plot(x1, r1(x1), linewidth=1.5)
plt.plot(x1, r2(x1), linewidth=1.5)

plt.plot(x1, z1(x1))
plt.plot(x1, z2(x1))
plt.plot(x1, z3(x1))
plt.plot(x1, z4(x1))
plt.plot(x1, z5(x1))

plt.ylim(0, 2)
plt.xlim(0, 2)

plt.show()
```

De acuerdo con la región definida por las restricciones y la posición de la función objetivo en el plano se pueden dar las siguientes situaciones:

*	Región factible no acotada.

*	Región factible vacía.

*	Infinitas soluciones.

*	Solución óptima única
